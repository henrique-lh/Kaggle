# Introduction

Oftentimes data will come to us with column names, index names, or other naming conventions that we are not satisfied with. In that case, you'll learn how to use pandas functions to change the names of the offending entries to something better.

```python
import pandas as pd
pd.set_option('max_rows', 5)
reviews = pd.read_csv('../input/wine-reviews/winemag-data-130k-v2.csv', index_col=0)
```

## Renaming

To rename a index names and/or column names, we can use the `rename()` method. You can specify if you're changing a column name or index name. This method takes a python dictionary as input, where the key is the old column name and the value will be the new column name. Let's see an example:

```python
reviews.rename(columns={'points': 'score'})
```

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

### Row and columns have their own names!

To set this, use the `rename_axis()` method

```python
reviews.rename_axis('wines', axis='rows').rename_axis('fields', axis='columns')
```

## Combine

We will sometimes need to combine different DataFrames/Series. There are a couple of methods to do this: `concat()`, `join()` and `merge()`. Let's see an example of how to use the `concat()` method

```python
canadian_youtube = pd.read_csv('../input/youtube-new/CAvideos.csv')
british_youtube = pd.read_csv('../input/youtube-new/GBvideos.csv')
```

```python
canadian_youtube.head(3)
```

```python
british_youtube.tail(3)
```

```python
pd.concat([canadian_youtube, british_youtube])
```

The `join()` method let's us combine different DataFrames objects which have an index in common

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```

The `lsuffix` and `rsuffix` parameters are necessary coz when we join the data, we wouldn't differenciate that, once the columns would have the same name
